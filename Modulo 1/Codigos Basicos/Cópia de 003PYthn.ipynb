```python
nome = input("digite seu nome ")
print(f"aeja bem vindo {nome}")
nota1 = float(input("digite sua primeira nota"))
nota2 = float(input("digite sua segunda nota"))
print(nota1 + nota2)
media = (nota1 + nota2)/2
print(f"o aluno {nome} teve nota de {media}")
```
