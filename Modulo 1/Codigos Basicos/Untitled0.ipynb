# Funçao print ()
A função print() é uma das funções mais importantes e usadas na linguagem Python. Sua funão é basicamente, exibir mensagens na tela ou envia las para outro dispositivo, como imprimir dentro de um arquivo de texto.

NO Python 3, print() é uma função interna, de modo que não é mecessário importar nenhuma biblioteca para poder utiliza-la. Basta chmá-la e passar os argumentos necessarios.


```python
print("meu primeiro print!") # todo txt deve estar entre aspas simples ou duplas.
```

```python
print('meu segundo print')
```

```python
print (D`água) # quando uma palavra contem aspas as aspas iniciais devem ser diferentes
```

```python
print(20+15)
```

```python
print("20+15") # o python realiza a operação que esta entro de parenteses
```

```python
print("20+15 = ", 20 + 15) # desta forma fica como uma conta feita.
```

```python
n1 = "1"
n2 = "3"
print(n1 + n2) #o resultado deu 1 pois o python entende tudo de forma literal 'string' (contatenar significa unir)
```

```python
nome = "Abed"
sobrenome = "sheikhi"
nome_completo = nome + " " + sobrenome
print(nome_completo)
```
