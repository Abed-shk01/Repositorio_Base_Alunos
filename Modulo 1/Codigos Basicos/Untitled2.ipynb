#Operadores lógicos

os operadores lógicos podem ser:
- AND (e)
- OR (ou)
- NOT (negação)

# AND

true + true = true  
true + false + false  
false + true = false  
false + false = false

```python
(10>20) and (20>30) #aqui diz que dois elementos falsos, o resultado e sempre falso.
```

```python
(10<20) and (20<30)
```

# OR

true + true = true  
true + false = true   
false + true = true    
false + false = false

```python
(1 < 2) or (3 > 6)
```

```python
'João' == 'João' or 'Python' != 'python' # isso deu true pois o primeiro e true e o segundo e falso
```

#NOT

not true = false or not false = true

```python
not True
```

```python
not False
```
