snake_case e um estilo de escrito usado para nome de variaveis, funções, omes em geral no codigo.
neste estilo as palavras sao todas em minisculas que separadas por um (-)

## Porque usar snake_case?


*   deixa o codigo mais facil de ler
*   E o padrão recomendado pela comunidade python

*   Segue a PEF8, que e guia de estilo oficial do pyhton

OBS: PEF8 é um documento que recomenda boas práticas pra escrever código Python limpo e legíve.
